Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the Dataset


```python
df = pd.read_csv('sales.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.reset_index(drop = True,inplace = True)
```

```python
df['city'] = df['Purchase Address'].str.split(',').str[1]

```

```python
df['state'] = df['Purchase Address'].str.split(',').str[2].str.split(' ').str[1]
```

```python
df
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'], errors='coerce').fillna(0).astype(int)
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'], errors='coerce').astype(float)
```

```python
df['sales'] = df['Quantity Ordered']*df['Price Each']
```

```python
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df.info()
```

```python
df['month'] = pd.to_datetime(df['Order Date']).dt.month
```

```python
df['Date'] = pd.to_datetime(df['Order Date']).dt.date
```

```python
df['year'] = pd.to_datetime(df['Order Date']).dt.year
```

```python
df
```

```python
df['Hour'] = pd.to_datetime(df['Order Date']).dt.hour
df['Minute'] = pd.to_datetime(df['Order Date']).dt.minute
df['Count'] = 1
```

```python
keys = [pair for pair, df in df.groupby(['Hour'])]
```

```python
df
```

```python

df.groupby('Product').sum()
```

```python
df = df[df['Product'] != 'Product']
```

```python

df.groupby('Product').sum()
```

```python
df.groupby('Product').mean()
```

# Which Product is sold the most

```python
product_group = df.groupby('Product')
```

```python
quantity_ordered =product_group.sum()['Quantity Ordered']

products = [product for product ,df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products,rotation='vertical',size=8)
plt.show()
```

# What Can be the reason behind it

```python
prices = df.groupby('Product').mean()['Price Each']
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products , prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
